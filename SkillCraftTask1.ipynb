```python
import pandas as pd # pandas library is used for data manipulation, reading, cleaning, transformation , integration etc.
import matplotlib.pyplot as plt # matplotlib library is used for visualising the data using different graphs
```

```python
df=pd.read_csv(r"C:\Users\ASUS\Downloads\indian_population_sample_100_unique.csv")# reading the csv file
print(df.head(50))# printing the top 50 rows/instances from the dataset
```

```python
if Age in df.columns:# iterating and finfing the age column in dataframe df columns
    plt.hist(df['Age'],bins=5,color="green",edgecolor="black")
    plt.title("Distribution of population across different age groups")#title for the histogram
    plt.xlabel("Age group Intervals")#label for X-axis
    plt.ylabel("Population")#label for Y-axis
    plt.grid(True,axis="y",linestyle='-', alpha=0.6)# showing grid lines 
    plt.show()#plotting the graph
else:
    print("Age column not found")
```

```python
# Example: if you're using a DataFrame named df we are calculating the value count for each unique value in Gender Column
gender_counts = df['Gender'].value_counts()

# Bar chart
plt.bar(gender_counts.index, gender_counts.values, color=['skyblue', 'lightpink'], edgecolor='black')

# Titles and labels
plt.title("Gender-wise Population Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of People")

# Adding value labels on top of bars for better understanding
for index, value in enumerate(gender_counts.values):
    plt.text(index, value + 1, str(value), ha='center')

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
